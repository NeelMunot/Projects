```python
def compare_list(lst5):
    lst2=lst5.copy()
    lst3=lst5.copy()
    lst2.sort()
    list3_1=list(set(lst3))
    list3_1.sort()
    if list3_1==lst2:
        return True
    else:
        return False
```
